# robustcov quickstart: robust distances

This notebook shows the core workflow: fit a robust covariance estimator, compute robust-distance anomaly scores, and inspect diagnostics.

```python
import numpy as np
import robustcov as rc

rng = np.random.default_rng(0)
X_clean = rng.normal(size=(500, 5))
X_out = rng.normal(loc=6.0, scale=1.5, size=(25, 5))
X = np.vstack([X_clean, X_out])
y = np.r_[np.zeros(len(X_clean), dtype=int), np.ones(len(X_out), dtype=int)]
```

```python
est = rc.FastMCD(random_state=0, quality="balanced").fit(X)
det = rc.RobustOutlierDetector(estimator=est, threshold="empirical", alpha=1 - y.mean()).fit(X)

pred = det.predict(X)
print("detected:", int(pred.sum()))
print("support:", int(est.support_.sum()))
print("radial kurtosis:", round(est.radial_kurtosis_, 3))
```

```python
report = rc.diagnostic_report(est, X)
print(report.summary())
```

The robust-distance profile and QQ plot are usually the first diagnostics to inspect. In scripts, pass `output_path=...` to save figures for docs or reports.

```python
rc.plot_robust_distance_profile(est, output_path="quickstart_distance_profile.png", show=False)
rc.plot_mahalanobis_qq(est, output_path="quickstart_qq.png", show=False)
```
