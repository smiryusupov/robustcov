# Small-sample heavy-tail scatter

This notebook demonstrates why `RegularizedCauchy` and `StudentTScatter` are useful when `p` is close to or larger than `n`.

```python
import numpy as np
import robustcov as rc

rng = np.random.default_rng(1)
n, p = 40, 60
A = rng.normal(size=(p, p))
true_cov = A @ A.T / p + 0.2 * np.eye(p)
L = np.linalg.cholesky(true_cov)
X = rng.standard_t(df=2, size=(n, p)) @ L.T
```

```python
cauchy = rc.RegularizedCauchy(alpha=0.10, warn_on_nonconvergence=False).fit(X)
student = rc.StudentTScatter(df=3, alpha=0.10, warn_on_nonconvergence=False).fit(X)
tyler = rc.RegularizedTyler(alpha=0.10).fit(X)

for name, est in [("Cauchy", cauchy), ("Student-t", student), ("Regularized Tyler", tyler)]:
    print(name, "cond=", round(est.condition_number_, 2), "radial_kurtosis=", round(est.radial_kurtosis_, 3))
```
