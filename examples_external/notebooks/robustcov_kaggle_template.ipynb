# robustcov Kaggle anomaly-detection template

Upload or attach a Kaggle CSV, set `DATA_PATH`, then run the cells. The notebook compares robustcov scores with common sklearn anomaly baselines and saves plots/metrics.

```python
DATA_PATH = '/kaggle/input/your-dataset/train.csv'
LABEL_COLUMN = 'target'
POSITIVE_LABEL = 1
OUTDIR = '/kaggle/working/robustcov_results'
```

```python
import sys
sys.path.append('/kaggle/working/robustcov_mvp/examples_external')
from _kaggle_utils import read_csv, numeric_matrix, scale_matrix, evaluate_external_baselines, save_table, print_table, plot_metric, plot_score_profile
from pathlib import Path

df = read_csv(DATA_PATH)
y = (df[LABEL_COLUMN].astype(str) == str(POSITIVE_LABEL)).astype(int).to_numpy()
contamination = max(float(y.mean()), 1.0 / len(y))
X, cols = numeric_matrix(df, label_col=LABEL_COLUMN)
X = scale_matrix(X)
rows, scores = evaluate_external_baselines(X, y, contamination=contamination, robust='auto')
print_table(rows, 'robustcov Kaggle benchmark')
```

```python
out = Path(OUTDIR)
save_table(rows, out / 'metrics.csv')
plot_metric(rows, 'pr_auc', out / 'pr_auc.png', 'PR AUC')
plot_metric(rows, 'f1', out / 'f1.png', 'F1 at contamination threshold')
plot_score_profile(scores, y, out / 'robust_score_profile.png', 'Robust score profile')
print('saved to', out)
```
